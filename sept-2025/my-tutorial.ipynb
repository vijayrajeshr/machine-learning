### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Import datasets

```python
file = pd.read_csv("insurance.csv")
file.head(3)
```

### Check for missing values

```python
file.isnull().sum()
file.head(2)
```

### Convert Text to Columns:
    # ML models cant directly work with text

```python
df = pd.get_dummies(file, drop_first= True)
df.head(3)
```

### Split Features as X and Target as Y
#-- Features X = age,sex,region(all the colums used to predict)
-- Target Y = We need to predict Cosr (the column that we want for PREDICTION) 

```python
x = df.drop('charges', axis=1) #axis 1 = "drop column" axis 0 = "drop row"
y = df['charges']
```

### Train-Test Split
-- Train on 80% data
-- Test on remaining 20% data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
```

### Training The model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train,y_train)
model
```

### Testing

```python
y_pred = model.predict(x_test)
```

### Evaluation

```python
from sklearn.metrics import mean_squared_error,r2_score

mse = mean_squared_error(y_test,y_pred)
root_mse = np.sqrt(mse)

r2 = r2_score(y_test,y_pred)


print("Mean Squared Error : ",mse)
print("Root Mean Squared Error : ",root_mse)
print("R^2 score : ",r2)
```
